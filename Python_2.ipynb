```python
df_estudiantes=estudiantes.sort_values('math score')
```

```python
#PARA DESCARGAR DATOS
df_estudiantes.to_csv('df_estudiantes.csv',index=False)
```

```python
#Organizar datos de manera descendente
df_estudiantes.sort_values('math score',ascending=False)
```

```python
#Ordenar dataframe por varias columnas. Primer dato es principal, segundo dependiente.
df_estudiantes.sort_values(['reading score','writing score'],ascending=False)
```

```python
organizado_lectura=df_estudiantes.sort_values(['reading score','writing score'],ascending=False)
organizado_lectura.to_csv('organizado_lectura.csv',index=False)
```

```python
#Guardar datos y sobreescribir sin usar la variable: m√©todo "inplace"
df_estudiantes.sort_values(['reading score','writing score'],ascending=False,
                           inplace=True)
```

```python
df_estudiantes
```

```python
#Organizar valores por texto: funcion "lambda"
#Convierte valores en minuscula y ordena
df_estudiantes.sort_values('race/ethnicity',ascending=True,
                           key=lambda col:col.str.lower())
```
