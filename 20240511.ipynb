```python
#Guardar raiz en una variable
ruta='https://www.football-data.co.uk/mmz4281/'
```

```python
#Guardar terminaciones/campeonatos en variables
campeonatos=['E0','E1','E2','E3','EC']
enlaces=[]

#Hay que recorrer lista con un loop (link dinamico)
#Una vez recorra la lista, debe reemplazar el valor para que el enlace funcione
for campeonato in campeonatos:
  df_link=pd.read_csv(ruta+'2425'+'/'+campeonato+'.csv')
  enlaces.append(df_link)
```
